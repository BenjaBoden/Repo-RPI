```python
import matplotlib.pyplot as plt
import numpy as np

X = [
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9],
]
X = np.array(X)
plt.scatter(X[:, 0], X[:, 1])
```

```python
mean = np.mean(X, axis=0)
B = X - mean
print(B)
plt.scatter(B[:, 0], B[:, 1])
```

```python
C = (1 / (len(B[:, 0]) - 1)) * B.T @ B
```

```python
D, V = np.linalg.eig(C)
print(D)
print(V)
```

```python
percentage = D / np.sum(D)
print(np.sum(D))
print(percentage)
```

```python
# Karhunen-Loeve Transforme
W = V[:, 1].reshape(2, 1)
KLT = B @ W
B_p = KLT @ W.T
print(B_p)
plt.scatter(B_p[:, 0], B_p[:, 1])
plt.scatter(B[:, 0], B[:, 1])
```
